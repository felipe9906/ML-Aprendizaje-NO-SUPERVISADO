# Seleccion nro de clusteres

## 1. Metodo del Codo
Se calcula y grafica la suma cuadrada en cada no de clusteres y se busca el cambio en la pendiente. En el caso de la imagen seria en 3. 


## 2. Metodo de la Silueta
Se debe establecer el procedimiento y tomar el punto mas alto de la grafica dada. Se da un intervalo hasta 1, entre mas cercano este a uno mejor será.


## 3. Metodo Estadistico de la Brecha
Se establece y da una grafica su punta sera la mas optima
